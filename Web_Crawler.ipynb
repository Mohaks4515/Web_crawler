```python
import requests
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup
from collections import deque

# Define the seed URL and depth
seed_url = "https://en.wikipedia.org/wiki/Animal_Kingdom_(TV_series)"
max_depth = 3

# Initialize the queue and visited sets
queue = deque([(seed_url, 0)])
visited = set()

# Define the BFS web crawler function
def bfs_web_crawler(seed_url,max_depth):
    while queue:
        url, depth = queue.popleft()
        if depth > max_depth:
            break
        if url in visited:
            continue
        visited.add(url)
        print("Crawling:", url)
        try:
            response = requests.get(url)
        except:
            continue
        soup = BeautifulSoup(response.text, 'html.parser')
        for link in soup.find_all('a'):
            href = link.get('href')
            if href is None:
                continue
            href = urljoin(seed_url, href)
            parsed_href = urlparse(href)
            if parsed_href.scheme not in ['http', 'https']:
                continue
            if parsed_href.netloc not in ['en.wikipedia.org']:
                continue
            if parsed_href.path.endswith(('.jpg', '.jpeg', '.gif', '.png', '.svg')):
                continue
            queue.append((href, depth + 1))

# Call the BFS web crawler function
bfs_web_crawler(seed_url,max_depth)
```
